```python
import asyncio
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
from scipy import signal
from scipy.signal import welch
from idun_guardian_sdk import GuardianClient

# --- KONFIGURATION ---
MY_API_TOKEN = "idun_MWSQ4pkewAGNz8wwYzw_NsweXihLC8tIcFzah8vqqys4Nc-ALzjfTwl2"
fs = 250 
WINDOW_SECONDS = 5 
BUFFER_SIZE = fs * WINDOW_SECONDS

# --- DATA STORAGE ---
class LiveData:
    def __init__(self):
        self.timestamps = []
        self.ch1 = []
        self.is_running = True

live_buffer = LiveData()

# --- FILTER SETUP (wie in deinem Original-Code) ---
sos_bp = signal.butter(N=4, Wn=[1, 40], btype="bandpass", fs=fs, output="sos")
b_notch, a_notch = signal.iirnotch(50, Q=30, fs=fs)

def apply_filters(data_array):
    if len(data_array) < fs:  # Warten, bis genug Daten für Filter da sind
        return data_array
    # Bandpass
    filtered = signal.sosfiltfilt(sos_bp, data_array)
    # Notch
    filtered = signal.filtfilt(b_notch, a_notch, filtered)
    # DC Offset entfernen
    filtered = filtered - np.mean(filtered)
    return filtered

# --- CALLBACK FUNKTION ---
def handle_live_data(event):
    # Filtern nach Raw-EEG Daten
    if type(event).__name__ == "LiveInsightsEvent":
        # Je nach SDK Struktur: event.message enthält oft die Liste der Samples
        # Falls event.message ein Dict mit 'raw_eeg' ist:
        if hasattr(event, 'message') and 'raw_eeg' in event.message:
            samples = event.message['raw_eeg']
            # Zeitstempel generieren oder vom SDK nehmen
            for sample in samples:
                live_buffer.ch1.append(sample)
                # Simulieren Timestamp, falls nicht im Event
                live_buffer.timestamps.append(len(live_buffer.ch1) / fs)
            
            # Buffer begrenzen, damit Notebook nicht abstürzt
            if len(live_buffer.ch1) > BUFFER_SIZE * 2:
                live_buffer.ch1 = live_buffer.ch1[-BUFFER_SIZE:]
                live_buffer.timestamps = live_buffer.timestamps[-BUFFER_SIZE:]

async def live_plot_task():
    """Diese Funktion aktualisiert den Plot alle 500ms"""
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))
    
    while live_buffer.is_running:
        if len(live_buffer.ch1) > fs:
            # Daten für Plot vorbereiten
            data = np.array(live_buffer.ch1[-BUFFER_SIZE:])
            times = np.array(live_buffer.timestamps[-BUFFER_SIZE:])
            
            # Filtern
            filtered_data = apply_filters(data)
            
            # Plot 1: Zeitbereich
            ax1.clear()
            ax1.plot(times, filtered_data, color='blue')
            ax1.set_title("Live EEG (Filtered)")
            ax1.set_ylabel("Amplitude (µV)")
            
            # Plot 2: PSD (Welch)
            freqs, psd = welch(filtered_data, fs=fs, nperseg=fs*2)
            ax2.clear()
            ax2.semilogy(freqs, psd)
            ax2.set_xlim([0.5, 40])
            ax2.set_ylim([0.01, 100])
            ax2.set_title("Power Spectrum (Live)")
            ax2.grid(True)
            
            clear_output(wait=True)
            display(fig)
            plt.pause(0.1) # Kurze Pause für das UI
            
        await asyncio.sleep(0.5)

async def main():
    client = GuardianClient(api_token=MY_API_TOKEN, debug=False)
    
    # Subscribe
    client.subscribe_live_insights(raw_eeg=True, handler=handle_live_data)
    
    print("Verbindung wird aufgebaut...")
    
    # Starte den Plot-Task im Hintergrund
    plot_task = asyncio.create_task(live_plot_task())
    
    try:
        # Aufnahme starten
        await client.start_recording(recording_timer=600, led_sleep=False)
    except Exception as e:
        print(f"Fehler: {e}")
    finally:
        live_buffer.is_running = False
        await plot_task
        print("Aufnahme beendet.")

# Start in Jupyter
await main() # Falls in Python File: asyncio.run(main())
```
